```python
import pandas as pd
```

```python
df=pd.read_excel("경기도 화장실 데이터.xlsx")
df
```

```python
df=df[~df['WGS84경도'].isna()&~df['WGS84위도'].isna()&~df['화장실명'].isna()]
df
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model
```

```python
import numpy as np
model.fit(np.column_stack([df['WGS84경도'], df['WGS84위도']]),df['화장실명'])
```

```python
import joblib

# 모델을 파일로 저장
joblib.dump(model, 'toiletmodel.pkl')
```
